```python
# Selecionei pelo dash apenas os dados que vou usar na análise 
# Porque se fosse carregar tudo aqui no jupyter ficaria muito pesado
# Para isso listo as colunas de dados que vou precisar e um faço um filtro 
# Pego esses dados processados e passo para formato .txt, tudo com o comando:
# cut -d'|' -f1,3,5,12,13,21,56,101,102,105 SUP_ALUNO_2019.CSV>exp_italo.txt

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('exp_italo.txt', sep="|", header=0)
```

```python
fig, ax = plt.subplots()

ax.hist(df[df["IN_CONCLUINTE"]==1]["TP_COR_RACA"], bins=15, alpha=0.5, color="blue", label="Concluiu")
ax.hist(df[df["IN_CONCLUINTE"]==0]["TP_COR_RACA"], bins=15, alpha=0.5, color="green", label="Não Concluiu")

ax.set_xlabel("Raça")
ax.set_ylabel("Total de pessoas")

fig.suptitle("Raça vs. Total de Concluintes")

ax.legend();
```

```python
df= df[df['NU_ANO_INGRESSO'] >= 2009]
df['IN_RESERVA_ETNICO'] = df['IN_RESERVA_ETNICO'].fillna(0)
# Tinham dados desde 1975, então limitei para a partir de 2009
```

```python
di_adm = {
1: "Pública Federal",
2: "Pública Estadual",
3: "Pública Municipal",
4: "Privada com fins lucrativos",
5: "Privada sem fins lucrativos",
6: "Privada - Particular em sentido estrito",
7: "Especial",
8: "Privada comunitária",
9: "Privada confessional"}

di_raca = {
0: "Aluno não quis declarar cor/raça",  
1: "Branca",
2: "Preta",
3: "Parda",
4: "Amarela",
5: "Indígena",
9: "Não dispõe da informação"}

df = df.replace({"TP_CATEGORIA_ADMINISTRATIVA": di_adm, "TP_COR_RACA": di_raca})
```
